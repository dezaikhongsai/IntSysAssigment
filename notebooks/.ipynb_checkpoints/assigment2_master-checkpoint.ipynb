```python
# === Cell 1: Load & Cleaning (Present steps + code) ===
import numpy as np
import pandas as pd

# 1) Load
df = pd.read_csv('diabetes.csv')
print("=== Raw info ===")
df.info()
print("\nHead:\n", df.head())

# 2) Kiểm tra null & 0
print("\n=== Null counts ===")
print(df.isnull().sum())

print("\n=== Counts of zeros (các cột có 0 là missing) ===")
zero_cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
print(df[zero_cols].eq(0).sum())

# 3) Quy ước: 0 -> NaN cho các cột y tế (vì 0 không hợp lý sinh học)
df[zero_cols] = df[zero_cols].replace(0, np.nan)

# 4) Bản copy để vẽ tương quan (fill tạm mean để không NaN khi vẽ)
df_corr_view = df.copy()
df_corr_view = df_corr_view.fillna(df_corr_view.mean(numeric_only=True))

# 5) Thống kê sau khi thay 0 -> NaN
print("\n=== Null counts after 0->NaN ===")
print(df.isnull().sum())

# Ghi chú:
# - Impute (điền khuyết) sẽ làm chuẩn trong Pipeline ở Cell 2 (KNNImputer) để tránh leakage khi CV.
```

```python
# === Cell 2: 5 features + 5 models + Evaluate with metrics ===
import numpy as np
import pandas as pd

# Nếu Cell 1 chưa chạy, tự nạp và xử lý tối thiểu
try:
    df
except NameError:
    df = pd.read_csv('diabetes.csv')
    zero_cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
    df[zero_cols] = df[zero_cols].replace(0, np.nan)

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, mean_absolute_error,
                             mean_squared_error, f1_score, roc_auc_score)

# 1) Chọn 5 features (nâng từ 3 lên 5)
features = ['Glucose','BMI','Age','BloodPressure','DiabetesPedigreeFunction']
X = df[features]
y = df['Outcome']

# 2) Định nghĩa 5 models "basic ML"
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "K Nearest Neighbors": KNeighborsClassifier(n_neighbors=11),  # có thể tune thêm
    "SVM Linear Kernel":   SVC(kernel='linear', probability=True, class_weight='balanced'),
    "SVM RBF Kernel":      SVC(kernel='rbf', probability=True, class_weight='balanced'),
    "Random Forest":       RandomForestClassifier(n_estimators=300, random_state=42)
}

# 3) Pipeline: Impute (KNN) + Scale (cho model cần) + Model
def make_pipe(estimator, need_scale=True):
    steps = [('imputer', KNNImputer(n_neighbors=5))]
    if need_scale:
        steps.append(('scaler', StandardScaler()))
    steps.append(('model', estimator))
    return Pipeline(steps)

pipes = {
    "Logistic Regression": make_pipe(models["Logistic Regression"], need_scale=True),
    "K Nearest Neighbors": make_pipe(models["K Nearest Neighbors"], need_scale=True),
    "SVM Linear Kernel":   make_pipe(models["SVM Linear Kernel"], need_scale=True),
    "SVM RBF Kernel":      make_pipe(models["SVM RBF Kernel"], need_scale=True),
    "Random Forest":       make_pipe(models["Random Forest"], need_scale=False),  # tree không cần scale
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 4) Đánh giá
rows = []
for name, pipe in pipes.items():
    acc = cross_val_score(pipe, X, y, cv=cv, scoring='accuracy').mean()
    f1  = cross_val_score(pipe, X, y, cv=cv, scoring='f1').mean()
    roc = cross_val_score(pipe, X, y, cv=cv, scoring='roc_auc').mean()

    # Tính MAE/MSE/RMSE trên nhãn 0/1 (theo yêu cầu đề), ý nghĩa hạn chế với classification
    y_pred_cv = cross_val_predict(pipe, X, y, cv=cv, method='predict')
    mae = mean_absolute_error(y, y_pred_cv)
    mse = mean_squared_error(y, y_pred_cv)
    rmse = np.sqrt(mse)

    rows.append([name, acc, mae, mse, rmse, f1, roc])

results_df = pd.DataFrame(rows, columns=["Model","Accuracy","MAE","MSE","RMSE","F1","ROC-AUC"])
results_df = results_df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)
print("\n=== Evaluation (sorted by Accuracy) ===\n", results_df)

best_name = results_df.iloc[0]["Model"]
best_pipe = pipes[best_name]
print(f"\nBest model by Accuracy: {best_name}")
```

```python
# === Cell 3: 5 visualizations ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Phòng hờ: nếu Cell 1/2 chưa chạy, tự tái tạo tối thiểu
try:
    df
except NameError:
    df = pd.read_csv('diabetes.csv')
    zero_cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
    df[zero_cols] = df[zero_cols].replace(0, np.nan)

try:
    df_corr_view
except NameError:
    df_corr_view = df.copy().fillna(df.copy().mean(numeric_only=True))

try:
    features
except NameError:
    features = ['Glucose','BMI','Age','BloodPressure','DiabetesPedigreeFunction']

from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, roc_curve, precision_recall_curve

X = df[features]
y = df['Outcome']

# Nếu best_pipe chưa có, dựng nhanh theo cách Cell 2 và chọn best
try:
    best_pipe, best_name
except NameError:
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
        "K Nearest Neighbors": KNeighborsClassifier(n_neighbors=11),
        "SVM Linear Kernel":   SVC(kernel='linear', probability=True, class_weight='balanced'),
        "SVM RBF Kernel":      SVC(kernel='rbf', probability=True, class_weight='balanced'),
        "Random Forest":       RandomForestClassifier(n_estimators=300, random_state=42)
    }
    def make_pipe(estimator, need_scale=True):
        steps = [('imputer', KNNImputer(n_neighbors=5))]
        if need_scale: steps.append(('scaler', StandardScaler()))
        steps.append(('model', estimator))
        return Pipeline(steps)
    pipes = {
        "Logistic Regression": make_pipe(models["Logistic Regression"], need_scale=True),
        "K Nearest Neighbors": make_pipe(models["K Nearest Neighbors"], need_scale=True),
        "SVM Linear Kernel":   make_pipe(models["SVM Linear Kernel"], need_scale=True),
        "SVM RBF Kernel":      make_pipe(models["SVM RBF Kernel"], need_scale=True),
        "Random Forest":       make_pipe(models["Random Forest"], need_scale=False),
    }
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    accs = {name: cross_val_score(pipe, X, y, cv=cv, scoring='accuracy').mean()
            for name, pipe in pipes.items()}
    best_name = max(accs, key=accs.get)
    best_pipe = pipes[best_name]

plt.rcParams["figure.figsize"] = (6, 6)

# 1) Correlation heatmap
plt.figure(figsize=(8,8))
sns.heatmap(df_corr_view[features + ['Outcome']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

# fit best model để vẽ các biểu đồ còn lại
best_pipe.fit(X, y)

# 2) Confusion Matrix (dùng OOF)
y_pred_oof = cross_val_predict(best_pipe, X, y, cv=5, method='predict')
cm = confusion_matrix(y, y_pred_oof)

plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cbar=False)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title(f"Confusion Matrix - {best_name}")
plt.tight_layout()
plt.show()

# 3) ROC Curve
y_proba = best_pipe.predict_proba(X)[:, 1]
fpr, tpr, _ = roc_curve(y, y_proba)
plt.figure()
plt.plot(fpr, tpr, label=f"{best_name}")
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curve"); plt.legend()
plt.tight_layout()
plt.show()

# 4) Precision-Recall Curve
prec, rec, _ = precision_recall_curve(y, y_proba)
plt.figure()
plt.plot(rec, prec, label=f"{best_name}")
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title("Precision-Recall Curve"); plt.legend()
plt.tight_layout()
plt.show()

# 5) Feature Importance (RandomForest)
rf_pipe = Pipeline([
    ('imputer', KNNImputer(n_neighbors=5)),
    # RF không cần scaler
    ('model', RandomForestClassifier(n_estimators=300, random_state=42))
])
rf_pipe.fit(X, y)
rf_model = rf_pipe.named_steps['model']
importances = rf_model.feature_importances_

plt.figure()
plt.barh(features, importances)
plt.xlabel("Importance"); plt.title("Feature Importance (Random Forest)")
plt.tight_layout()
plt.show()
```

```python
# === Cell 4: Save pipeline & demo prediction ===
import numpy as np
import pandas as pd
import pickle
import os

# Phòng hờ: nếu best_pipe chưa có, thử load file đã lưu; nếu chưa có nữa thì dựng nhanh
try:
    best_pipe, best_name
except NameError:
    if os.path.exists("diabetes_pipeline.sav"):
        best_pipe = pickle.load(open("diabetes_pipeline.sav", "rb"))
        best_name = "Loaded from file"
    else:
        # Dựng nhanh:
        df = pd.read_csv('diabetes.csv')
        df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = \
            df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0, np.nan)
        features = ['Glucose','BMI','Age','BloodPressure','DiabetesPedigreeFunction']
        X = df[features]
        y = df['Outcome']

        from sklearn.pipeline import Pipeline
        from sklearn.impute import KNNImputer
        from sklearn.preprocessing import StandardScaler
        from sklearn.svm import SVC

        best_name = "SVM RBF Kernel"
        best_pipe = Pipeline([
            ('imputer', KNNImputer(n_neighbors=5)),
            ('scaler', StandardScaler()),
            ('model', SVC(kernel='rbf', probability=True, class_weight='balanced'))
        ])
        best_pipe.fit(X, y)

# Lưu pipeline đầy đủ
save_path = "diabetes_pipeline.sav"
with open(save_path, "wb") as f:
    pickle.dump(best_pipe, f)
print(f"Saved: {save_path} (model = {best_name})")

# Load lại để demo
loaded_pipe = pickle.load(open(save_path, "rb"))

# Demo input (5 features, đúng thứ tự & tên cột)
demo = pd.DataFrame([[65, 70, 50, 72, 0.5]],
                    columns=['Glucose','BMI','Age','BloodPressure','DiabetesPedigreeFunction'])

pred = loaded_pipe.predict(demo)[0]
proba = loaded_pipe.predict_proba(demo)[0, 1]
conf = round(max(proba, 1-proba)*100, 2)

print("Prediction:", "Diabetic" if pred == 1 else "Non-diabetic")
print("Confidence:", f"{conf}%")
```
